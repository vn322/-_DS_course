Посчитайте значение первой и второй производной функции:f(x) = x^5 + 4sin(2x) + cos(3x+3)


В точке x = 1.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
```

```python
x= sym.Symbol('x')

foo1 = x**5 + 4*sym.sin(2*x) + sym.cos(3*x+3)

diff1 = sym.diff(foo1, x, 1)
diff2 = sym.diff(diff1)
foo1
```

```python
diff1
```

```python
diff2
```

```python
diff1.subs(x,1)
```

```python
diff2.subs(x,1)
```

Постройте график первой и второй производной функции:\f(x) = sin^5(2x+1)


На отрезке [-5; 5].

```python
foo2 = sym.sin(2*x+1)**5
foo2
```

```python
dif1 = sym.diff(foo2)
dif2 = sym.diff(dif1)
dif1
```

```python
from sympy.plotting import plot

plot(dif1,dif2, ('x', -5, 5))
```
