Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},

перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
Общие требования к программе:

# Задание 1
Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
Примеры работы:

1.

Введите команду:
p

Введите номер документа:
10006
Результат:
Владелец документа: Аристарх Павлов

2.

Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def FunP():
    
    name = ''
    number = input('Введите номер документа: ')
    for i in documents:
        if i['number'] == number:
            name = 'Владелец документа: ' + i['name']
    if name == '': 
        name = 'Документ не найден в базе'
    return name
```

Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
Примеры работы:

1.

Введите команду:
s

Введите номер документа:
10006
Результат:
Документ хранится на полке: 2

2.

Введите команду:
s

Введите номер документа:
12345
Результат:
Документ не найден в базе

```python
def FunS():
    
    shelf = ''
    number = input('Введите номер документа: ')
    for item, values in directories.items():
        if number in values:
            shelf = 'Документ хранится на полке: ' + item
    if shelf == '':
        shelf='Документ не найден в базе'
    return shelf
```

Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
Пример работы:

Введите команду:
l
Результат:

№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

```python
def FunL():
    
    for i in documents:
        for item, values in directories.items():
            if i['number'] in values:
                print('№: ' + i['number'] + ', тип: ' + i['type'] + ', владелец: ' + i['name'] + ', полка хранения: ' + item)
```

Пункт 4. Пользователь по команде "ads" может добавить новую полку
Примеры работы:

1.

Введите команду:
ads

Введите номер полки:
10
Результат:
Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

2.

Введите команду:
ads

Введите номер полки:
1
Результат:
Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

```python
def shelf():

    x = ''
    for item, values in directories.items():
        x = x + item + ',' 
    x = x[:-1] + '.'
    return x
```

```python
def FunAds():

    number = input('Введите номер полки: ')
    if number in directories:
      print('Такая полка уже существует. Текущий перечень полок: ' + shelf())

    else:
        directories[number]=[]
        print('Полка добавлена. Текущий перечень полок: '+ shelf())
```

Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
Примеры работы:

1.

Введите команду:
ds

Введите номер полки:
3
Результат:
Полка удалена. Текущий перечень полок: 1, 2.

2.

Введите команду:
ds

Введите номер полки:
1
Результат:
На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

3.

Введите команду:
ds

Введите номер полки:
4
Результат:
Такой полки не существует. Текущий перечень полок: 1, 2, 3.

```python
def FunDs():

    number = input('Введите номер полки: ')
    if number in directories:
      if directories[number] != []:
        print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ' + shelf())
      else:
        del directories[number]
        print('Полка удалена. Текущий перечень полок: '+ shelf())

    else:
        print('Такая полка не существует. Текущий перечень полок: ' + shelf())
```

```python
def main():

    x = 0
    while x < 1:
        command_input=input('Введите команду: ')
        if command_input == 'p':           
            print(FunP())
        elif command_input == 's':
            print(FunS())
        elif command_input == 'l':
            FunL()
        elif command_input == 'ads':
            FunAds()
        elif command_input == 'ds':
            FunDs()
        elif command_input == 'q':
            break
        else:
            print('Нет такой команды')
```

```python
main()
```
