```python
df_sport = df[df['Genre'] == 'Sports']
df_sport.head()
```

```python
df_sport.describe()
```

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(df_sport['Critic_Score'], 70)

print(df_sport['Critic_Score'].mean())
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

```python
df_pc = df[df['Platform'] == 'PC'][['Critic_Score']]
df_pc
```

```python
df_ps = df[df['Platform'] == 'PS4'][['Critic_Score']]
df_ps
```

```python
st.ttest_ind(df_pc, df_ps, equal_var=False)
```

p-value меньше 0.01 - значимые различия

```python
import nltk
#from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords_set = set(stopwords.words('english'))
texts_splitted = [text.split() for text in df['Message']]
without_sw = [[word for word in text if word not in stopwords_set] for text in texts_splitted]
```

```python
df['Message'] = [' '.join(row) for row in lemmatized_text]
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = tfidf_matrix 
y = df['Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```
