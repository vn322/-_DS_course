# Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

# Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, model.predict(X_train)) # Оцениваем качество на выборке для обучения
```

```python
mean_squared_error(y_test, model.predict(X_test))# Оцениваем качество на выборке для тестирования RSE
```

```python
mean_squared_error(y_test, model.predict(X_test), squared=False)# Оцениваем качество на выборке для тестирования RMSE
```

```python
 z = ['unknown', 'Action', 'Adventure',
       'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
       'Sci_Fi', 'Thriller', 'War', 'Western']
df_av = [] 
 for i  in z:
   
   df_av += (df_spark.groupby([i]).avg('rating').collect()[1])
```

```python
av = df_av[1::2]
av
```

```python
av = list(zip(z, av))
av = sorted(av, key=lambda x: (x[1]), reverse = True)
av
```
